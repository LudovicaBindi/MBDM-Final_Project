```python
from ema_workbench.analysis import parcoords

from problem_formulation import get_model_for_problem_formulation

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.simplefilter("ignore")
```

In this book we show the code we use to print <code>Parallel coordinates plots</code> for the results of our experimentats (steps 2 and 6) with the policies found from the optmizations steps (steps 1 and 5).

### Printing the result of the first optimization (step 1)

```python
# read the data
results = pd.read_csv('intermediate outputs/step1 - first optimization output(100000,[0.1]).csv')
# set up the model
dike_model, planning_steps = get_model_for_problem_formulation(6)
```

```python
# visualization in the parallel coordinate plots

# we are keeping just the columns with the outcomes that we minimized in the optimization step
data = results.loc[:, [o for o in ['A.4_Dike Investment Costs', 'A.4_Expected Annual Damage',
       'A.4_Expected Number of Deaths', 'A.5_Dike Investment Costs',
       'A.5_Expected Annual Damage', 'A.5_Expected Number of Deaths',
       'Expected Evacuation Costs', 'Other.Dikes_Expected Annual Damage',
       'Other.Dikes_Total Dike Investment Costs',
       'Other.Dikes_Expected Number of Deaths']]]

limits = parcoords.get_limits(data) # creates a dataframe where for each outcome it gets the highest and lowest value

paraxes = parcoords.ParallelAxes(limits) # creates the parallel axes

paraxes.plot(data) # put the data on the axes
plt.rcParams["figure.figsize"]=(15,7) # fix the size of the picture

# save figure
plt.savefig('output pictures/step1 - results first optimization.png')

# plots
plt.show() 
```

### Printing the result of the second optimization (step 5)

```python
# read the data
results = []
for i in range(3):
    results.append(pd.read_csv('intermediate outputs/step5 - multi scenario optimization results/step5 - ' + str(i) + ' optimization results.csv'))
    
results = pd.concat(results)

# set up the model
dike_model, planning_steps = get_model_for_problem_formulation(6)
```

```python
# visualization in the parallel coordinate plots

# we are keeping just the columns with the outcomes that we minimized in the optimization step
data = results.loc[:, [o for o in ['A.4_Dike Investment Costs', 'A.4_Expected Annual Damage',
       'A.4_Expected Number of Deaths', 'A.5_Dike Investment Costs',
       'A.5_Expected Annual Damage', 'A.5_Expected Number of Deaths',
       'Expected Evacuation Costs', 'Other.Dikes_Expected Annual Damage',
       'Other.Dikes_Total Dike Investment Costs',
       'Other.Dikes_Expected Number of Deaths']]]

limits = parcoords.get_limits(data) # creates a dataframe where for each outcome it gets the highest and lowest value

paraxes = parcoords.ParallelAxes(limits) # creates the parallel axes

paraxes.plot(data) # put the data on the axes
plt.rcParams["figure.figsize"]=(20,10) # fix the size of the picture

# save figure
plt.savefig('output pictures/step6 - results second optimization.png')

plt.show() # plots
```

This picture is not very meaningful (e.g., too many colors for the policy). Let's try highlight the policies that perform best safety-wise to get more insights in the trade-offs that the policy makers need to make.

```python
# create a copy of the original results
results = results.reset_index()
temp = results.copy(deep=True)
```

```python
# find the 'best' policies by looking at the distribution of the results over the outcomes of interest
deaths_perc_4 = np.percentile(temp['A.4_Expected Number of Deaths'], 30)
temp = temp.loc[temp['A.4_Expected Number of Deaths'] < deaths_perc_4]
deaths_perc_5 = np.percentile(temp['A.5_Expected Number of Deaths'], 30)
len(temp.loc[temp['A.5_Expected Number of Deaths'] < deaths_perc_5])
```

```python
# create an array that can be used to select the policies from the the original dataframe
y = []
to_keep = temp.index.tolist()
for index, row in results.iterrows():
    if (row['A.4_Expected Number of Deaths'] < deaths_perc_4) & (row['A.5_Expected Number of Deaths'] < deaths_perc_5):
        y.append(True)
    else:
        y.append(False)
```

```python
np.sum(y)
```

```python
# plot the policies
data = pd.DataFrame({k:v[y] for k,v in results.items()}) # the policies we want to highlight
all_data = pd.DataFrame({k:v for k,v in results.items()}) # all the policies

# plot the policies over our client's main outcomes of interest, safety ones
selected_data = all_data[['A.4_Expected Number of Deaths','A.5_Expected Number of Deaths','Other.Dikes_Expected Number of Deaths']]

# create the parallel coordinate plot
limits = parcoords.get_limits(selected_data)
plt.rcParams["figure.figsize"] = (20,10)
axes = parcoords.ParallelAxes(limits)

axes.plot(all_data, color='lightgrey')
axes.plot(data, color='blue') # highlight the policies we are interested in

# save figure
plt.savefig('output pictures/step6 - results second optimization - lowest deaths dike rings of clients.png')

plt.show() # plots
```
